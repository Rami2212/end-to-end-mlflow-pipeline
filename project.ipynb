```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
import mlflow
import mlflow.sklearn
from sklearn.datasets import load_diabetes
```

## Data preparation

```python
diabetes = load_diabetes()
print(diabetes)
```

```python
df = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)
df['bmi'] = diabetes.target
```

```python
df
```

```python
# independent and dependent variables
X = df.drop(labels=['bmi'], axis=1)
y = df['bmi']
```

```python
# model
def hyperparameter_tuning(X_train, y_train, param_grid):
    rf = RandomForestRegressor()
    grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=3, n_jobs=-1, verbose=1, scoring='neg_mean_squared_error')
    grid_search.fit(X_train, y_train)
    return grid_search
```

```python
# train test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# infersignature
from mlflow.models import infer_signature

signature = infer_signature(X_train, y_train)

# params
params = {
    'n_estimators': [100, 200],
    'max_depth': [5, 10],
    'min_samples_split': [5, 10, None],
    'min_samples_leaf': [2, 5]
}
```

```python
import mlflow
import mlflow.sklearn
from urllib.parse import urlparse
from mlflow.models import infer_signature
from sklearn.metrics import mean_squared_error

# set tracking URI
mlflow.set_tracking_uri("http://127.0.0.1:5000")

with mlflow.start_run():
    grid_search = hyperparameter_tuning(X_train, y_train, params)

    # best model
    best_model = grid_search.best_estimator_

    # predict
    y_pred = best_model.predict(X_test)

    # evaluate
    mse = mean_squared_error(y_test, y_pred)

    # log best parameters
    mlflow.log_param("best_n_estimators", grid_search.best_params_['n_estimators'])
    mlflow.log_param("best_min_samples_split", grid_search.best_params_['min_samples_split'])
    mlflow.log_param("best_min_samples_leaf", grid_search.best_params_['min_samples_leaf'])
    mlflow.log_param("best_max_depth", grid_search.best_params_['max_depth'])

    # log metric
    mlflow.log_metric("mse", mse)

    # log model
    tracking_get_uri = urlparse(mlflow.get_tracking_uri()).scheme
    if tracking_get_uri != "file":
        mlflow.sklearn.log_model(best_model, "model", registered_model_name="best_regression_model")
    else:
        signature = infer_signature(X_train, y_pred)
        mlflow.sklearn.log_model(best_model, signature=signature)

    # always print results
    print(f"best hyperparameters: {grid_search.best_params_}")
    print(f"mean squared error: {mse}")
```
